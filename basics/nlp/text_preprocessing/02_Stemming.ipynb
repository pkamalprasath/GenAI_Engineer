### Stemming
Stemming is like cutting a word to its smallest root—just like cutting extra leaves from a plant to see the stem! We remove the ending part of the word so we can see the main meaning.So even if the words look different—played, playing, player—they all go back to the same stem: “play”.

1.PorterStemmer

```python
words=["playing", "played", "plays","jumping", "jumped", "runs", "running","leaves", "wolves"]
```

```python
#function import
from nltk.stem import PorterStemmer 
#Object declaration of the method
stemming=PorterStemmer() 
```

```python
for word in words:
    print(word+"---->"+stemming.stem(word))
```

### 2.RegexpStemmer Class 
Regex stemming is like having magic scissors that follow special patterns to cut the endings of words.
Instead of cutting every word the same way, you tell the scissors a rule, like:

“Cut -ing if you see it at the end!” ; “Cut -ed if it’s there!” ;“Cut -s only if it is plural!” 
These rules are written using regex, which is like a small language for matching patterns.

```python
# Importing the function 
from nltk.stem import RegexpStemmer
# creating the object along with required regrex input parameter
# Stemming will work based on the regex patter
# $ sign indicates where the stemming should happen
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

### Snowball Stemmer 
Snowball Stemmer is like a super-smart word-cutting robot.It knows many languages and has better rules than normal stemmers.If stemming is like cutting leaves off a plant,then Snowball is the gardener who cuts neatly, cleanly, and correctly!It removes endings like:–ing,–ed,–ly,–es,–ness,–ment:so all words become their simple root.

```python
from nltk.stem import SnowballStemmer 
# we need to select the language that we are using
snowball=SnowballStemmer('english')
```

```python
for word in words: 
    print(word+"---->"+snowball.stem(word))
```
