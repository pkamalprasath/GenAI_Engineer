```python
corpus = """Do you think customers will accept a millet celebration cake? 
Last weekend, five customers ordered again—no preservatives."""
```

```python
print(corpus)
```

```python
## Tokenization
## Paragragh --> Sentence 

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize
documents=sent_tokenize(corpus)
```

```python
documents
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
# Paragragh --> words


from nltk.tokenize import word_tokenize 
words=word_tokenize(corpus) 
words 
```

```python
# Sentence --> words 
for sentence in documents: 
    print(word_tokenize(sentence)
    )    
```
