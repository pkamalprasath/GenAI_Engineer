### Lemmatizer 
Is like having a smart dictionary friend who knows the real, correct form of every word.
Instead of just cutting the word (like stemming), it changes the word to its proper dictionary form.#### Simple difference
Stemming = cutting  he word to a short root (might look weird)

Application - Q & A , Text summarization ; Cons : its is time consumming 

```python
words=["playing", "played", "plays","jumping", "jumped", "runs", "running","leaves", "wolves"] 
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer 
```

```python
lemmatizer=WordNetLemmatizer()
lemmatizer.lemmatize("sentences")
```

```python
'''POS - Noun - n 
Verb - v 
adjective - a 
adverb -r '''
```

```python
lemmatizer.lemmatize("looking",pos='v')
```

```python
for word in words: 
    print(word+"---->"+lemmatizer.lemmatize(word,pos='n'))
```
